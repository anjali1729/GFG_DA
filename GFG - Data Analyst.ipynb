```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("GFG Data Analyst test.csv", skiprows=2)
```

```python
df.drop(df.tail(7).index,
        inplace = True)
```

```python
df.columns
```

```python
del df['Unnamed: 7']
del df['Unnamed: 8']
```

```python
df["Spend in $"] = pd.to_numeric(df["Spend in $"])
df["Visits"] = pd.to_numeric(df["Visits"].str.replace(',', ''))
df["Orders"] = pd.to_numeric(df["Orders"].str.replace(',', ''))
df["Revenue in $"] = pd.to_numeric(df["Revenue in $"].str.replace(',', ''))
df["Week"] = pd.to_numeric(df["Week"])
```

```python
# Conversion Rate (CR) = Orders/Visits (%)
df['Conversion Rate'] = df['Orders']*100/df['Visits']
# Average Basket Size (ABS) = Revenue/Orders ($)
df['Average Basket Size'] = df['Revenue in $']/df['Orders']
# Cost to Income Ratio (CIR) = Spend/Revenue ($)
df['Cost to Income Ratio'] = df['Spend in $']/df['Revenue in $']
```

```python
df
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='Week', y='Conversion Rate', data=df, hue='Channel name')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='Week', y='Average Basket Size', data=df, hue='Channel name')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='Week', y='Cost to Income Ratio', data=df, hue='Channel name')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=90)
plt.show()
```
